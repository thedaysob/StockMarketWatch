```python
# Class that groups stocks together
class portfolio:
    """
    Portfolio is made up of assets and cash equivilants.
    """
    def __init__(self, cashEq = 5000):
        """
        Construct a Portfolio object.
        
        :param cashEQ: [float] - cash equivilant is USD
        :param assets: [dict] - dictionary in form {asset-name[str] : quantity owned[float]}
        :param assetVal: [float] - asset value in USD
        :param 
        :param start: [dateTime64(yyyy-mm-dd)] - start date for trading universe
        :param end: [dateTime64(yyyy-mm-dd)] - end date for trading universe
        """
        
        self.cashEq = cashEq
        self.assets = {}
        self.assetVal = 0
        self.portVal = self.cashEq + assetVal
        self.start = yf.Ticker('TSLA').history('5y').index[0]
        self.end = yf.Ticker('TSLA').history('5y').index[-1]
        
    
    def depositCash(self, deposit):
        """
        Deposit cash equivilant into Portfolio. 
        
        :param deposit: [float] - USD value of deposited cash
        """
        self.cashEq += deposit
        self.portVal += deposit
        
    
    def withdrawCash(self, withdrawl):
        """
        Withdraw cash equivilant from Portfolio. 
        
        :param withdrawl: [float] - USD value of withdrawn cash
        """
        self.cashEq -= withdrawl
        self.portVal -= withdrawl
        
    
    def buyAsset(self, tickers, quantities, date):
        """
        Add asset to Portfolio.
        
        :param tickers: [list(str)] - valid tickers of assets to be purchased
        :param quantities: [list(int)] - quantity of asset to be purchased 
        :param date: [dateTime64(yyyy-mm-dd)] - date of asset purchase
        """
        
        #Check each asset has corresponding quantity to purchase.
        if len(tickers) != len(quantity):
            print("each asset much have corresponding quantity")
            return 
        
        for i in range(len(tickers)):
            try:
                ticker = yf.Ticker(tickers[i])
                quant = quantities[i]
                
                assetName = ticker.info['shortName']
                assetPrice = ticker.history().loc[date].Close
                
                if ticker in self.assets:
                    self.assets[ticker] += quant
                else:
                    self.assets[ticker] = quant
                
                self.assetVal += assetPrice * quant
                self.cashEq -= assetPrice * quant
                
                print("Purchased: " +  quant + " " + assetName + "assets for " + assetPrice * quant)
                print("Value of Portfolio assets: " + self.assetVal)
                print("Value of Portfolio cash equivilant: " + self.cashEq)
            
            except:
                print(tickers[i] + " is not a valid asset symbol")
                
                
    def sellAsset(self, tickers, quantities, date = self.start):
        """
        Add asset to Portfolio.
        
        :param tickers: [list(str)] - valid tickers of assets to be sold
        :param quantities: [list(int)] - quantity of asset to be sold 
        :param date: [str(yyyy-mm-dd)] - date of asset sale
        """
        
        #Check each asset has corresponding quantity to purchase.
        if len(tickers) != len(quantity):
            print("each asset much have corresponding quantity")
            return 
        
        for i in range(len(tickers)):
            try:     
                ticker = yf.Ticker(tickers[i])
                quant = quantities[i]
                
                assetName = ticker.info['shortName']
                assetPrice = ticker.history().loc[date].Close
                
                if ticker in self.assets:
                    self.assets[ticker] -= quant
                else:
                    print("You must own " + assetName + " asset in order to sell it")
                    return
                
                self.assetVal -= assetPrice * quant
                self.cashEq += assetPrice * quant
                
                print("Sold: " + quant + " " + assetName + " assets for " + assetPrice * quant)
                print("Value of Portfolio assets: " + self.assetVal)
                print("Value of Portfolio cash equivilant: " + self.cashEq)
            
            except:
                print(tickers[i] + " is not a valid asset symbol")
                
    
    def createTimeSeriesDF(self, start = self.start, end = self.end)
         """
        Create time-series dataFrame of cashEq value, each asset value, and portfolio value from start date to end date.
        
        :param start: [dateTime64(yyyy-mm-dd)] - start date for time-series
        :param end: [dateTime64(yyyy-mm-dd)] - end date for time-series
        """
        
        arr = np.array()
        
        return 0
        
        
    # Param:
    #     period: time duration
    # Functionality:
    #     Graphs the price at closing for all stocks within the stocklist for specified amount time
    
    
    
    
    def graphStocksOnClose(self, period):
        if period not in ['1d', '5d', '1mo', '3mo', '6mo', '1y', '2y', '5y', '10y', 'ytd', 'max']:
            print(period + " is not available period")
            return
        
        colorIndex = 0
        shortNameList = []
        for stock in self.stockList:
            tickerClient = yf.Ticker(stock)
            data = tickerClient.history(period)
            info = tickerClient.info
            print(info['shortName'])
            priceOnClose = data['Close']
            plt.plot(priceOnClose, label=info['shortName'])
                        
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

#     # Param:
#     #     period: time duration
#     # Functionality:
#     #     Graphs the stock price percentage change from the previous day for all stocks within the stocklist for specified amount time
#     def graphStocksByPercentChange(self, period):
#         if period not in ['1d', '5d', '1mo', '3mo', '6mo', '1y', '2y', '5y', '10y', 'ytd', 'max']:
#             print(period + " is not available period")
#             return
        
#         colorIndex = 0
#         percentChangeList = []
        
#         for stock in self.stockList:
#             tickerClient = yf.Ticker(stock)
#             data = tickerClient.history(period)
#             info = tickerClient.info
#             print(info['shortName'])
#             priceOnClose = data['Close']
#             length = len(priceOnClose)
            
#             percentChangeList = priceOnClose.copy()
#             for i in range(1,length):
#                 priceChange = priceOnClose[i] - priceOnClose[i-1]
#                 percentChangeList[i] = priceChange / priceOnClose[i-1]
#             percentChangeList[0] = 0
#             plt.plot(percentChangeList, label=info['shortName'])
#             plt.xticks(rotation=90)

#         plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
        
        
#     def sectorBreakdown(self)
    
#     def returnOnInv
        
        
    
    
        
```

```python
def rollingAvgDF(stock, period):
        if period not in ['1y', '2y', '5y', '10y', 'ytd', 'max']:
            print(period + " is not available period")
            return
        stock = yf.Ticker(stock)
        data = stock.history(period)
        name = stock.info['shortName']
        data['rolling20Avg'] = data.loc[:,'Close'].rolling(window=20).mean()
        data['rolling50Avg'] = data.loc[:,'Close'].rolling(window=50).mean()
        rollingAvgDF = data[['Close','rolling20Avg', 'rolling50Avg']]
        return name, rollingAvgDF
    
def graphRollingAvg(stock, period):
    name, rollAvgDF = rollingAvgDF(stock, period)
    plt.plot(rollAvgDF['Close'], label='close')
    plt.plot(rollAvgDF['rolling20Avg'], label='20dAvg')
    plt.plot(rollAvgDF['rolling50Avg'], label='50dAvg')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.title(name + "Close & Rolling Avgs")
```
